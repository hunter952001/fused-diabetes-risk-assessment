```python
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_auc_score, classification_report

# Input the precision on the POSITIVE CLASS from the two models
# Positive class because we are concerned with predicting diabetes
# We would rather have false positives
Precision_CNN_1 = 0.98
Precision_RF_1 = 0.31
Recall_CNN = 0.87
Recall_LR = 0.79
F1_CNN = 0.92
F1_LR = 0.45

import numpy as np

# Load saved logits (make sure Drive is mounted)
logits_cnn = np.load('/content/drive/MyDrive/MLMA Final Project/Rough Project Outline/logits_cnn_ft.npy')
logits_lr = np.load('/content/drive/MyDrive/MLMA Final Project/Rough Project Outline/logits_lr.npy')

# Select a random index from each
logit_cnn_sample = np.random.choice(logits_cnn)
logit_lr_sample = np.random.choice(logits_lr)

# Convert logits to probabilities using sigmoid
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

p_cnn = sigmoid(logit_cnn_sample)
p_lr = sigmoid(logit_lr_sample)

# Compute fusion weight alpha
alpha = F1_CNN / (F1_CNN + F1_LR)

# Final fused risk score
final_risk = alpha * p_cnn + (1 - alpha) * p_lr

# Print everything for clarity
print(f"Random CNN logit: {logit_cnn_sample:.4f} → P_cnn = {p_cnn:.4f}")
print(f"Random LR  logit: {logit_lr_sample:.4f} → P_lr  = {p_lr:.4f}")
print(f"Fusion weight α  = {alpha:.2f}")
print(f"Final fused diabetes risk score: {final_risk:.4f}")
```
